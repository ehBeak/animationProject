```python
#video capture_com
vidcap = cv.VideoCapture('com.mp4')
success,image_bgr = vidcap.read()

count = 1
success = True

while success:
  success,image_bgr = vidcap.read()
  cv.imwrite("/content/imgcom/%d.png" % count, image_bgr)
  print("saved imagecom %d.png" % count)
  
  if cv.waitKey(10) == 27:                    
      break
  count += 1
```
